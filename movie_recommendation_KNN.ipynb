```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neighbors import NearestNeighbors

# Load dataset
data = pd.read_csv("../admin1/Documents/tmdb_5000_credits.csv")

data = data[['title','cast','crew']]

# Handle missing values
data['cast'] = data['cast'].fillna('')
data['crew'] = data['crew'].fillna('')

# Combine features
data['tags'] = data['cast'] + " " + data['crew']

# Vectorization
vectorizer = CountVectorizer(stop_words='english')
matrix = vectorizer.fit_transform(data['tags'])

# KNN model
model = NearestNeighbors(metric='cosine', algorithm='brute')
model.fit(matrix)

def recommend(movie):

    index = data[data['title'] == movie].index[0]

    distances, indices = model.kneighbors(matrix[index], n_neighbors=6)

    print("Recommended Movies:\n")

    similarity_scores = []

    for i, d in zip(indices[0][1:], distances[0][1:]):
        title = data.iloc[i]['title']
        similarity = (1 - d) * 100
        similarity_scores.append(similarity)

        print(title, "- Similarity:", round(similarity,2), "%")

    # Average similarity (pseudo accuracy)
    avg_accuracy = sum(similarity_scores) / len(similarity_scores)
    print("\nModel Recommendation Accuracy:", round(avg_accuracy,2), "%")

# Test
recommend("Avatar")
```

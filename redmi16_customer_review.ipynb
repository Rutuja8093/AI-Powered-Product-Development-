```python
import pandas as pd

df = pd.read_csv("../admin1/Downloads/redmi6.csv", encoding="latin1")
df
```

```python
# Convert rating to numeric (1–5)
df['rating_num'] = df['Rating'].str[0].astype(int)
df
```

```python
def rating_to_sentiment(r):
    if r >= 4:
        return "Positive"
    else:
        return "Negative"

```

```python
df['Sentiment'] = df['rating_num'].apply(rating_to_sentiment)

df.head()
```

```python

X = df['Comments'].astype(str)
y = df['Sentiment']

from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from sklearn.feature_extraction.text import TfidfVectorizer

custom_stopwords = list(ENGLISH_STOP_WORDS - {'not', 'no', 'never'})

tfidf = TfidfVectorizer(
    stop_words=custom_stopwords,
    ngram_range=(1,2)
)



X_tfidf = tfidf.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    class_weight='balanced',  # balances Positive/Negative automatically
    max_iter=1000
)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

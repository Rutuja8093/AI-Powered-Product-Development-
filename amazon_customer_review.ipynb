```python

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].str.replace('%',''), errors='coerce')
```

```python
def rating_to_sentiment(r):
    if r >= 4:
        return "Positive"
    else:
        return "Negative"

```

```python
df['Sentiment'] = df['rating'].apply(rating_to_sentiment)

df.head()
```

```python
# Count of each sentiment
sentiment_counts = df['Sentiment'].value_counts()
print(sentiment_counts)
```

```python

X = df['review_content'].astype(str)
y = df['Sentiment']

from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from sklearn.feature_extraction.text import TfidfVectorizer

custom_stopwords = list(ENGLISH_STOP_WORDS - {'not', 'no', 'never'})

tfidf = TfidfVectorizer(
    stop_words=custom_stopwords,
    ngram_range=(1,2)
)



X_tfidf = tfidf.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model1 = LogisticRegression(
    class_weight='balanced',  # balances Positive/Negative automatically
    max_iter=1000
)

model1.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model1.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
